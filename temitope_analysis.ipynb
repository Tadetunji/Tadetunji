```python
import pandas as pd
```

```python
df= pd.read_csv('Retail Store Sales.csv')
```

```python
df.head(10)
```

```python
df.columns
```

```python
df['final_pay']= df['Total Sales (₦)']* df['Discount (%)']
```

```python
df.head()
```

```python
df2 = pd.pivot_table(df, values='Total Sales (₦)',index= ['Category','Product'],aggfunc= ['sum', 'count','mean' ])
```

```python
print(df2)
```

```python
df.head()
```

```python
df['new_class'] = df.apply(lambda row: 'Cheap' if row['Unit Price (₦)'] <= 5000 else ('Expensive' if row['Unit Price (₦)'] <= 10000 else 'very expensive'), axis=1)

```

```python
df.head()
```

```python
df3 = pd.pivot_table(df, values='Total Sales (₦)',index= ['new_class','Category','Product'],aggfunc= ['count'])
```

```python
print(df3)
```

```python
df3.dropna(axis= 1)
```

```python
print(df3)
```

```python
df4 = pd.pivot_table(df, values='Total Sales (₦)',index= ['Category','Product'],columns= 'new_class', aggfunc= ['count'],)
```

```python
print(df4)
```

```python
df4.fillna(df4.mean(), inplace= True)
```

```python
print(df4)
```

```python
df4.to_csv('pivot_retail.csv')
```
